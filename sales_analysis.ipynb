```python
#importando

import pandas as pd
import numpy as np 

df_sales = pd.read_csv("sales.csv")

```

```python
#padronizando

df_sales.columns = df_sales.columns.str.lower().str.strip()

df_sales.columns

df_sales["date"] = pd.to_datetime(df_sales["date"])


```

```python
#verificando estado dos dados

df_sales.isnull().sum()

df_sales.info()

df_sales.describe()
```

```python
#colunas estrategicas 

df_sales["month"] = df_sales["date"].dt.to_period("M")
df_sales["weekday"] = df_sales["date"].dt.day_name()
```

```python
#metricas

total = df_sales["total_amount"].sum()

ticket = df_sales["total_amount"].mean()

qtd = df_sales["quantity"].sum() 

```

```python
#vendas ao longo do tempo

df_sales.groupby("month")["total_amount"].sum()
```

```python
#Categoria mais lucrativa

df_sales.groupby("product_category")["total_amount"].sum().sort_values(ascending=False)
```

```python
#recorrentes vs novos 

df_sales.groupby("is_returning_customer")["total_amount"].mean()
```

```python
#dispositivos

df_sales.groupby("device_type")["total_amount"].mean()
```

```python
#Entrega vs satisfação

df_sales.groupby("delivery_time_days")["customer_rating"].mean()
```

```python
#Tempo no site vs compra

df_sales.groupby("session_duration_minutes")["total_amount"].mean()
```

```python
df_sales["session_group"] = pd.cut(
    df_sales["session_duration_minutes"],
    bins=[0,10,20,30],
    labels=["curto","medio","longo"]
)

df_sales.groupby("session_group")["total_amount"].mean()
```

# Insights principais

- Clientes novos apresentam ticket médio ligeiramente superior, porém a diferença é pequena.
- Clientes recorrentes tendem a gerar maior valor total ao longo do tempo.
- A maior parte das sessões ocorre entre 12 e 16 minutos, com comportamento de compra estável.
- Sessões mais longas apresentam maior valor médio, mas com baixa frequência (possíveis outliers).
- O tipo de dispositivo influencia o comportamento de compra.

# Conclusão

O dataset apresenta um comportamento consistente de vendas, com forte influência de fatores como tempo de sessão, tipo de cliente e categoria de produto.

Esses insights podem ser utilizados para:
- melhorar estratégias de marketing
- otimizar experiência do usuário
- aumentar retenção de clientes
